# Property Price Estimator

## Library imports

```python
import pandas as pd
import numpy as np
```

## Load data

```python
df_retail = pd.read_excel('../data/properties_retail.xlsx')
df_resell = pd.read_excel('../data/properties_resell.xlsx')
```

```python
# Adding new binary variable to distinguish between retail and resell markets
df_retail['market'] = 0
df_resell['market'] = 1
```

```python
df = pd.concat([df_retail, df_resell], ignore_index=True)
```

```python
# Set float display format to plain decimal
pd.set_option('display.float_format', '{:.2f}'.format)
```

## Clean data

```python
df.isna().sum()
```

```python
# Target Variable creation
df['price_per_square_meter'] = df['price'] / df['surface']
```

```python
# Drop unnecessary columns and the columns that create target variable
df = df.drop(columns=['id', 'url', 'title', 'location', 'sewage', 'heating', 'price'])
# I was not sure if 'surface' should be dropped as well, but I decided to keep it as it could be useful for further analysis and modeling.
# I was concerned that it might create a data leakage, but surface is aways available and it is not derived from price, so I believe it should be fine to keep it.
```

```python
# Drop rows with missing critical features (except price)
df = df.dropna(subset=['rooms', 'floor','construction_year', 'latitude', 'longitude', 'surface'])
```

```python
# Fill remaining missing values with 'inny'
df['material'] = df['material'].fillna('inny')
```

```python
df.isna().sum()
```

```python
# Convert data types
numeric_columns = {
    'price_per_square_meter': 'float',
    'surface': 'float',
    'rooms': 'Int64',
    'floor': 'Int64',
    'construction_year': 'Int64',
    'latitude': 'float',
    'longitude': 'float',
    'market': 'Int64'
}
df = df.astype(numeric_columns)
```
